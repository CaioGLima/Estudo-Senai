```python
portuguese_columns = {
    'Unnamed: 0': 'Sem Nome: 0',
    'Accident_Index': 'Índice_Acidente',
    'Location_Easting_OSGR': 'Localização_Easting_OSGR',
    'Location_Northing_OSGR': 'Localização_Northing_OSGR',
    'Longitude': 'Longitude',
    'Latitude': 'Latitude',
    'Police_Force': 'Força_Policial',
    'Accident_Severity': 'Gravidade_Acidente',
    'Number_of_Vehicles': 'Número_Veículos',
    'Number_of_Casualties': 'Número_Feridos',
    'Date': 'Data',
    'Day_of_Week': 'Dia_da_Semana',
    'Time': 'Hora',
    'Local_Authority_(District)': 'Autoridade_Local_(Distrito)',
    'Local_Authority_(Highway)': 'Autoridade_Local_(Rodovia)',
    '1st_Road_Class': 'Classe_1ª_Estrada',
    '1st_Road_Number': 'Número_1ª_Estrada',
    'Road_Type': 'Tipo_Estrada',
    'Speed_limit': 'Limite_Velocidade',
    'Junction_Control': 'Controle_Cruzamento',
    '2nd_Road_Class': 'Classe_2ª_Estrada',
    '2nd_Road_Number': 'Número_2ª_Estrada',
    'Pedestrian_Crossing-Human_Control': 'Travessia_Pedestres-Controle_Humano',
    'Pedestrian_Crossing-Physical_Facilities': 'Travessia_Pedestres-Instalações_Físicas',
    'Light_Conditions': 'Condições_Luminosidade',
    'Weather_Conditions': 'Condições_Climáticas',
    'Road_Surface_Conditions': 'Condições_Superfície_Estrada',
    'Special_Conditions_at_Site': 'Condições_Especiais_Local',
    'Carriageway_Hazards': 'Riscos_Pista',
    'Urban_or_Rural_Area': 'Área_Urbana_ou_Rural',
    'Did_Police_Officer_Attend_Scene_of_Accident': 'Policial_Presente_Local_Acidente',
    'LSOA_of_Accident_Location': 'LSOA_Local_Acidente',
    'Year': 'Ano'
}
```

```python
df.rename(columns= portuguese_columns ,inplace=True)
#rename ele vai renomear o nome das colunas. Inplace ele aplica a alteração
```

```python
df.drop(columns=['Sem Nome: 0','Condições_Especiais_Local','Riscos_Pista','LSOA_Local_Acidente','Riscos_Pista','Condições_Especiais_Local',],inplace=True)
```

```python
df[df['Controle_Cruzamento'].isna()]
```

```python
df['Controle_Cruzamento'] = df['Controle_Cruzamento'].fillna('')
#Codigo responsavel por substituir os valores null por "espaço" e joga novamente na coluna
#o fillna só vai considerar valores null, os outros valores ele ignora
```

```python
df['Limite_Velocidade'] = df['Limite_Velocidade'].fillna(df['Limite_Velocidade'].mean())
#aqui ele é o codigo de cima, mas ele passou um valor no fillna, que agora vai ser a média
```

```python
df['Hora'] = df['Hora'].fillna(df['Hora'].mode()[0])
```

```python
df['Travessia_Pedestres-Instalações_Físicas'] = df['Travessia_Pedestres-Instalações_Físicas'].fillna('non-confirmation')
#aqui ele vai criar uma coluna para os valores null e nomea ela
```

```python
df['Travessia_Pedestres-Controle_Humano'] = df['Travessia_Pedestres-Controle_Humano'].fillna('non-confirmation')
```

```python
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True, errors="coerce")
df['Hora'] = pd.to_datetime(df["Hora"], format="%H:%M", errors="coerce").dt.hour
#indicam o formato que essas tabelas tem que trabalhar com os valores delas
```

```python
#df["Ano"] = df["Data"].dt.year
df["Mes"] = df["Data"].dt.month
df["Nome_Mes"] = df["Data"].dt.month_name()
```

```python
pd.set_option('display.max_columns', None)
#retira o limite de linhas que aparecem
display(df[0:5])
#indica a quantidade que pode aparece
```

```python
map_dias = {
    1: "Domingo",
    2: "Segunda",
    3: "Terça",
    4: "Quarta",
    5: "Quinta",
    6: "Sexta",
    7: "Sábado"
}

df["Dia_Semana_Nome"] = df["Dia_da_Semana"].map(map_dias)
#Antes do = eh oq vc cria;
#Depois do = ja ta no conjunto de dados
```

```python
map_gravidade = {
    1: "Fatal",
    2: "Grave",
    3: "Leve",
    9: "Desconhecida"
}

df["Gravidade_Label"] = df["Gravidade_Acidente"].map(map_gravidade)
#Antes do = eh oq vc cria;
#Depois do = ja ta no conjunto de dados
```

```python
df["Escuro"] = df["Condições_Luminosidade"].str.contais("Dark", case=False, na=False)
df["Chuva"] = df["Condições_Climáticas"].str.contais("Rain", case=False, na=False)
```
