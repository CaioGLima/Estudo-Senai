```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/DataSet/StudentsPerformance.csv')
```

```python
print(df.columns)
```

```python
translation_dict = {
    'gender': 'gênero',
    'race/ethnicity': 'raça/etnia',
    'parental level of education': 'nível de escolaridade parental',
    'lunch': 'almoço',
    'test preparation course': 'curso de preparação para o teste',
    'math score': 'pontuação em matemática',
    'reading score': 'pontuação em leitura',
    'writing score': 'pontuação em escrita'
}
#Dicionario com colunas traduzidas
```

```python
df.rename(columns= translation_dict ,inplace=True)
#renomeção das colunas
```

```python
#criação de uma nova coluna, chamada "Pontuação Total"
df['pontuação total'] = df['pontuação em matemática'] + df['pontuação em leitura'] + df['pontuação em escrita']
df.head()
```

```python
#mostra os 5 primeios head(5) e 5 ultimos (tail)
display(df)
```

```python
#para ver as informações
df.info()
```

```python
print(df.isnull().sum())
#verificação de null, não possue nenhum vazio
```

#Tipos de tratamentos que podemos fazer:

#Quantidade de gêneros
#Quantidade de gêneros por etnia
#Top estudantes
#Comparação de top homens x mulheres nas matérias

#Value counts na coluna raça etnia, curso de preparação e lvl de escolaridade média por pontuação


```python
#Quantidade de Generos

df['gênero'].value_counts()
```

```python
#Quantidade de Generos por etnia
quant_Gen_Et = df.groupby(['gênero','raça/etnia',]).size().unstack(fill_value=0)
quant_Gen_Et
```

```python
#Top 10 estudantes pra cima
df_tabela = df.sort_values(by='pontuação total', ascending=False).reset_index(drop=True)
top_3_melhores = df_tabela.head(3)
top_3_melhores
```

```python
#Top 10 estudantes pra baixo
df_tabela = df.sort_values(by='pontuação total', ascending=True).reset_index(drop=True)
top_3_piores = df_tabela.head(3)
top_3_piores
```

```python
#Quantidade de melhores alunos por etnia
top_melhores = top_3_melhores['raça/etnia'].value_counts()
top_melhores
```

```python
#Quantidade de piores alunos por etnia
top_piores = top_3_piores['raça/etnia'].value_counts()
top_piores
```

```python
#Comparação numero de H x M no top 100 melhores
df_tabela = df.sort_values(by='pontuação total', ascending=False).reset_index(drop=True)
top_100_estudantes = df_tabela.head(100)
top_100 = top_100_estudantes['gênero'].value_counts()
top_100
```

```python
#agrupamento de raça, lvl de escolaridade e curso em relação a pontuação média total das matérias
df.groupby(['raça/etnia', 'nível de escolaridade parental', 'curso de preparação para o teste'])['pontuação total'].mean().reset_index().round(2)
```
