Com base na tabela abaixo, escreva um programa que leia o código de um item e a quantidade deste item. A seguir, calcule e mostre o valor da conta a pagar.



```python
#criar uma lista, acessa a lista com os valores e faz a soma

Codigo= [1,2,3,4,5]
Nome= ["Cachorro Quente", "X-Salada","X-Bacon","Torrada Simples","Refrigerante"]
Valor= [4.0, 4.50, 5.0, 2.0, 1.50]

print("Cardapio")
for i in range(len(Codigo)):
  print(f"{Codigo[i]} - {Nome[i]} - {Valor[i]}")

codigo = int(input("Informe o codigo do produto: "))
quant = int(input("Insira a quantidade do produto: "))

if codigo in Codigo:
  total = Valor[codigo - 1] * quant
  print("Preço a pagar R$:", total)
else:
  print("Codigo invalido")
```
