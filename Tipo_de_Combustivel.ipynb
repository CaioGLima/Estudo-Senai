Um Posto de combustíveis deseja determinar qual de seus produtos tem a preferência de seus clientes. Escreva um algoritmo para ler o tipo de combustível abastecido (codificado da seguinte forma: 1.Álcool 2.Gasolina 3.Diesel 4.Fim). Caso o usuário informe um código inválido (fora da faixa de 1 a 4) deve ser solicitado um novo código (até que seja válido). O programa será encerrado quando o código informado for o número 4.

```python
Alcool = 0
Gasolina = 0
Diesel = 0

while True:
  print("----Menu----")
  print("1. Alcool")
  print("2. Gasolina")
  print("3. Diesel")
  print("4. Fim")

  opcao = int(input("Qual operação deseja fazer: "))


  if opcao == 1:
    Alcool = int(input("Informe quantas pessoas abasteceram hoje: "))

  elif opcao == 2:
    Gasolina = int(input("Informe quantas pessoas abasteceram hoje: "))

  elif opcao == 3:
    Diesel = int(input("Informe quantas pessoas abasteceram hoje: "))

  elif opcao ==4:
    print("Alcool:", Alcool)
    print("Gasolina:",Gasolina)
    print("Diesel:",Diesel)
    print("Muito obrigado")
    break

  else:
    print("Opcao errada, insira um valor valido")
    break

```
